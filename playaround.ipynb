```python
driver_path = "/usr/local/bin/chromedriver"
# Set up the WebDriver
service = Service(driver_path)
driver = webdriver.Chrome(service=service)
```

```python
search_text = "Location:"
```

```python
job_divs = driver.find_elements(By.XPATH, f"//div//*[contains(text(), '{search_text}')]//ancestor::div[1]")
```

Extract disciplines

```python
divs = driver.find_elements(By.XPATH, "//div[a]")
```

```python
links = [div.find_element(By.TAG_NAME, "a").get_attribute('href') for div in divs]
links
```

```python
disciplines_links = [link for link in links if 'search' in link and 'all-disciplines' not in link and 'hub' not in link]
disciplines_links
disciplines_links
```

```python
disciplines_divs = []
links_set = set(links)
for i in range(len(links)):
    if links[i] in disciplines_links:
        disciplines_divs.append(divs[i])
```

```python
disciplines_array = np.unique(np.array([div.text for div in np.array(disciplines_divs)]))
disciplines_array
```
